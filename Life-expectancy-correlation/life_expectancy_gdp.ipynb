```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('all_data.csv')
df.head(20)
```

```python
df = df.rename(columns={'Life expectancy at birth (years)': 'life_expectancy'})

df_Chile = df[df.Country == 'Chile']
df_Zimbabwe = df[df.Country == 'Zimbabwe']
df_USA = df[df.Country == 'United States of America']
df_China = df[df.Country == 'China']
df_Germany = df[df.Country == 'Germany']
df_Mexico = df[df.Country == 'Mexico']
countries = list(df.Country.unique())
countries_dict = {}
for i in countries:
    countries_dict[i] = df[df.Country == i]

    
```

```python
sns.histplot(df['life_expectancy'],bins=20)

plt.show()
```

Most of the values for life expectancy in the table are between 75 and 85 years. Zimbabwe is visible outlier, citizens of this country born between 2000 and 2015 have life expectancy between 45 and 60 years. The distribution is left skewed with some small amount of values on the left of the histogram.

```python
sns.lmplot(x='life_expectancy',y='GDP',data=df)
plt.show()
```

The gdp seem to be colerated with the life expectancy. Let's check how strong the correlation is.

```python
correlation, p = pearsonr(df.life_expectancy,df.GDP)
print(correlation)
```

It seems that the correlation is lower than expected. The 0.34 correlation means that it is pretty weak. Lets double check how does it look for each country separetly.

```python
def calc_correlation(dataframe,column1,column2,country):
    corr,p = pearsonr(dataframe[column1],dataframe[column2])
    print('Correlation for ' + country +' is ' + str(corr))
```

```python
for k in countries_dict:
    calc_correlation(countries_dict[k],'life_expectancy','GDP',k)
```

As we can see if we calculate the correlation separately for each country we concluded that the correlation between gdp and life expectancy is very strong. The closer to 1 is the value, the stronger the correlation.

```python
countries = list(df.Country.unique())

df_mean = df.groupby(['Country'])['life_expectancy'].mean()
list_mean_exp = list(df_mean)
plt.figure(figsize=(10,10))
ax = plt.subplot(1,1,1)
barlist = plt.bar(countries,list_mean_exp)
ax.set_xticks(range(len(countries)),color='white')
ax.set_xticklabels(['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'], color='white')
ax.set_yticks([i for i in range(81) if i % 10 == 0],color='white')
ax.set_yticklabels([i for i in range(81) if i % 10 == 0],color='white')
plt.ylabel('Average life expectancy at birth in years',color='white')
plt.xlabel('Country',color='white')
plt.title('Average life expectancy per country at birth for years 2000-2015',color='white')
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax = plt.subplot(1,1,1)
sns.lineplot(data=df_Chile,x='Year',y='life_expectancy',label='Chile')
sns.lineplot(data=df_China,x='Year',y='life_expectancy',label='China')
sns.lineplot(data=df_Mexico,x='Year',y='life_expectancy',label='Mexico')
sns.lineplot(data=df_USA,x='Year',y='life_expectancy',label='USA')
sns.lineplot(data=df_Germany,x='Year',y='life_expectancy',label='Germany')
sns.lineplot(data=df_Zimbabwe,x='Year',y='life_expectancy',label='Zimbabwe')
plt.xlabel('Year',color='white')
plt.ylabel('Life expectancy for birth year per country',color='white')
plt.legend()
plt.show()
```

All the analyzed countries have their life expectancy for the 15 year period on relatively stable and high levels from low 70s up to 80. In comparision Zimbabwe life expectancy have been on extremaly low levels for the time preiod and between 2000-2004 was even decreasing. In 2005 the trend reversed and over 2005-2015 period life expectancy in Zimbabwe grew from 45 to around 60 years.
